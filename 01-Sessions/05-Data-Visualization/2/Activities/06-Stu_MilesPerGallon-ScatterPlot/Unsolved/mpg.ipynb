```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
car_data = pd.read_csv('../Resources/mpg.csv')
car_data.head()
```

```python
# Remove the rows with missing values in horsepower
clean_car_data = car_data.loc[car_data['horsepower'] != "?"]
```

```python
clean_car_data
```

```python
# Set the 'car name' as our index
clean_car_data = clean_car_data.set_index('car name')
clean_car_data

```

```python
# Remove the 'origin' column
clean_car_data = clean_car_data.drop(columns=['origin'])
```

```python
clean_car_data.info()
```

```python
# Convert the "horsepower" column to numeric so the data can be used

clean_car_data["horsepower"] = pd.to_numeric(clean_car_data["horsepower"])
```

```python
clean_car_data.info()
```

```python
# Create a scatter plot which compares MPG to horsepower

clean_car_data.plot(x= 'horsepower', y= 'mpg' , kind="scatter")
```

```python
# Create a scatter plot which compares MPG to horsepower
clean_car_data.plot(kind="scatter", x="horsepower", y="mpg", grid=True, figsize=(8,8),
              title="MPG Vs. Horsepower",rot=45,fontsize=15)
plt.show()
```
